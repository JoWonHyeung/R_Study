```python
import chromedriver_autoinstaller
import time
import csv
import re
from selenium import webdriver
from bs4 import BeautifulSoup

broswer = webdriver.Chrome("C:\\tmp\\chromedriver.exe")
broswer.get("https://cafe.naver.com/zzang9daddy?iframe_url=/ArticleList.nhn%3Fsearch.clubid=29328371%26search.menuid=9%26search.boardtype=I")
time.sleep(1)
broswer.switch_to.frame("cafe_main") 
```

```python
count = 3
urls = []

#url ÏàòÏßë
while True:
    html = broswer.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    nextCss = "#main-area > div.prev-next > a:nth-child(" + str(count) + ")"
 
    all = soup.find('ul','article-album-sub').find_all('li')
    default_url = "https://cafe.naver.com/zzang9daddy.cafe"
  
    #url ÏàòÏßë
    for i in all:
        tag = i.find('a')
        urls.append(default_url + tag['href'])
    
    if count % 10 == 1:
        broswer.find_element_by_css_selector("#main-area > div.prev-next > a.pgR").click()
        count = 3
    else:
        broswer.find_element_by_css_selector(nextCss).click()
        count = count + 1
    
    time.sleep(2)
    print(len(urls))
    if len(urls) == 20000:
        break
```

```python
import pandas as pd

#df = DataFrame()
all_info = []
cnt = 0
flag = 0

def filter_func(x):
   if x == None or x == '\u200b':
     return None
   else:
     return x

#url Ï†ëÍ∑º ÌõÑ Ï†ïÎ≥¥ Ï∂îÏ∂ú

for url in urls[1500:]:
    broswer.get(url)
    time.sleep(2)
    broswer.switch_to.frame("cafe_main") 
    time.sleep(1)
    html = broswer.page_source
    soup = BeautifulSoup(html, 'html.parser')

    infos = soup.select('.se-main-container > .se-component > .se-component-content > .se-section > .se-module > .se-text-paragraph')

    for info in infos:
        info = re.sub('üîàÌï¥ÎãπÍ≤åÏãúÌåêÏùÄ ÏÉùÎ∞©ÏÜ° Ïª®ÌÖêÏ∏†Ïóê ÏÇ¨Ïö© Îê† Ïàò ÏûàÏäµÎãàÎã§.','',info.get_text())
        info = re.sub('Îç∞ÏùºÎ¶¨Î£© Í¥ÄÎ†® Í≥µÏßÄ ->https://cafe.naver.com/zzang9daddy/719128','',info)
        info = re.sub('<Í∏∞Î≥∏ÏñëÏãù>|ÌòëÏ∞¨|ÏÑ†ÌÉù','',info)
        info = re.sub('ÏïÑÏö∞ÌÑ∞|ÏÉÅÏùò|ÌïòÏùò|Ïã†Î∞ú|Í∏∞ÌÉÄ ÏïÖÏÑ∏|\s','',info)
        info = re.sub('ÌÇ§,Î™∏Î¨¥Í≤å','',info)
        info = re.sub('[()-@#!$%^&*]','',info)

        emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                                "]+", flags=re.UNICODE)
        info = emoji_pattern.sub(r'', info)
        
        all_info.append(info)
        
        all_info = list(filter(filter_func, all_info))
        
        
        if len(all_info) > 10000:
            flag = 1
            break
    print(len(all_info))
    time.sleep(1)
   
    if flag == 1:
        break
```
